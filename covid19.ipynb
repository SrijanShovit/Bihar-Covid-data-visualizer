```python
web_page.columns = ['districts','total cases','recoveries','deaths','active cases']
```

```python
web_page.drop([38],inplace=True)
```

```python
web_page
```

```python
tot = pd.Series(map(int,web_page['total cases']))
D = pd.Series(map(int,web_page['deaths']))
AC = pd.Series(map(int,web_page['active cases']))
R = pd.Series(map(int,web_page['recoveries']))
```

```python
fig = plt.figure(figsize=(15,5))
axes = fig.add_axes([0,0,1,1])
x = np.arange(38*4,step=4)
axes.bar(x,tot,width=0.8,label="Total Cases")

axes.bar(x+0.8,D,width=0.8,label="Deaths")
axes.bar(x+1.6,R,width=0.8,label="Recoveries")
axes.bar(x+2.4,AC,width=0.8,label="Active Cases",color="yellow")
plt.xticks(x+0.8,web_page['districts'],rotation=60)
axes.legend()
```
